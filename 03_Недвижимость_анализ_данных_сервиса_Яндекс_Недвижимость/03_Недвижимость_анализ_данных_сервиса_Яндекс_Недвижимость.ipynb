```python
# приведу названия всех столбцов к змеиному регистру. 
estate_data = estate_data.rename(
    columns={
        'cityCenters_nearest':'city_centers_nearest',
        'parks_around3000':'parks_around_3000',
        'ponds_around3000':'ponds_around_3000'
    }
)
# проверю, что названия изменились
estate_data.columns
```

Названия всех столбцов теперь оформлены единым стилем в змеином регистре. Теперь обработаем каждый столбец.

Большинство запросов для изучения и обработки данных в столбцах будут выглядеть однотипно, поэтому задам функции.

```python
# функция для просмотра информации по столбцу
def column_analisis (column):
    print('Количество пропущенных значений в столбце равно:', estate_data[column].isna().sum())
    print()
    print('Первые 5 значений столбца')
    print(estate_data[column].head())
    print()
    print(estate_data[column].describe())
    estate_data[column].hist(bins=100, figsize=(15, 6))
    plt.show()
    
```

```python
# функция для замены пропусков медианными значениями, сгруппированными по другому столбцу
# column - столбец, где необходимо заполнить пропуски
# gr_column - столбец, по которому группируются данные
def val_gr_median(column, gr_column):
    for r  in estate_data[gr_column].unique():
        estate_data.loc[(estate_data[gr_column] == r) & (estate_data[column].isna()), column] = \
        estate_data.loc[(estate_data[gr_column] == r), column].median()
```

```python
# функция для замены пропусков средними значениями, сгруппированными по другому столбцу
# column - столбец, где необходимо заполнить пропуски
# gr_column - столбец, по которому группируются данные
def val_gr_mean(column, gr_column):
    for r  in estate_data[gr_column].unique():
        estate_data.loc[(estate_data[gr_column] == r) & (estate_data[column].isna()), column] = \
        estate_data.loc[(estate_data[gr_column] == r), column].mean()
```

В названии населенных пунктов оставлю только название само название, без типа населенного пункта. Для этого список со значениями, которые необходимо удалить и функцию для удаления

```python
# Все населенные пункты, в названии которых есть "поселок", переименую просто в "поселок"
dict_replace_vill = ['городской поселок', 'городской посёлок', 'коттеджный поселок', 'коттеджный посёлок', 'поселок городского типа', 
                'поселок станции', 'посёлок', 'посёлок городского типа', 'посёлок при железнодорожной станции', 'посёлок станции', 
                'поселок при железнодорожной станции']

# Создаю функцию, которая удалит все значения, которые есть в словаре
for value in dict_replace_vill:
    estate_data['locality_name'] = estate_data['locality_name'].str.replace(value, 'поселок')

# Все садовые товарищества будут с типом населенного пункта "садовое товарищество"
dict_replace_gard = ['некоммерческое ']

# Создаю функцию, которая удалит все значения, которые есть в словаре
for value in dict_replace_gard:
    estate_data['locality_name'] = estate_data['locality_name'].str.replace(value, '')
    
# На всякий случай заменю все "ё" на "е"
estate_data['locality_name'] = estate_data['locality_name'].str.replace('ё', 'е', regex = True)

# Приведу все значения к нижнему регистру
estate_data['locality_name'] = estate_data['locality_name'].str.lower()


# Еще раз вывожу на экран список уникальных значений
print(estate_data['locality_name']
      .sort_values()
      .unique()
     )
```
