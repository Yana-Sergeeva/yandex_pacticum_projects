# Поиск закономерностей, определяющих успешность компьютерных игр

## Описание данных

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

### Анализ продаж на платформах

```python
# Изучу суммарные продажи на различных платформах.
platform_sales = (df
 .pivot_table(index='platform', values='total_sales', aggfunc='sum')
 .sort_values(by='total_sales', ascending=False)
 .reset_index()                 
 
)
platform_sales
```

```python
# Посмотрю на график
platform_sales.plot(title='Продажи на платформах', kind='bar', x='platform', y='total_sales', figsize=(15,6))
plt.xlabel('Платформа')
plt.ylabel('Продажи, млн копий')
```

По таблице и диаграмме видно, что начиная с платформы 2600, продажи сильно отличаются. Буду считать платформы с наибольшими суммарными продажами с индекса 0 до 17 включительно.

```python
# Отредактирую platform_sales
platform_sales = platform_sales.head(18)
platform_sales
```

```python
# Создам список с названиями платформ
top_platform_sales = platform_sales['platform'].to_list()
top_platform_sales
```

```python
# Теперь применю этот список для фильтрации основного дафафрейма и построю распределение продаж на этих платформах по годам
platform_year = df.query('platform in @top_platform_sales')
                 
(
    platform_year
    .pivot_table(index='year_of_release', columns='platform',values='total_sales',aggfunc='sum')
    .plot(figsize=(15,6))
)
```

Очень похоже на рисунки годовалого ребенка))) Но все же можно увидеть, что в период восходящего тренда выпуска игр с 1995 по 2009 появлялось и существовало большое количество платформ.

```python
# Посчитаю сроки существования платформ
platform_life = platform_year.groupby(['platform', 'year_of_release']).agg({'total_sales':'sum'}).reset_index()
platform_life['platform'].value_counts()
```

По этим данным видно, что большинство платформ существуют 10-12 лет.

### Определение актуального периода для составления прогнозов на 2017 год

Я определила, что большинство платформ существуют 11-12 лет. Следовательно, чтобы определить актуальный период надо взять данные за последние 12 лет, т.е. после 2005 года

```python
# Изучу количество выпускаемых игр на различных платформах с 2006 года
(
    df
    .query('year_of_release > 2005')
    .pivot_table(index='platform', columns='year_of_release', values='name', aggfunc='count')
)
```

По этой таблице видно, что с 2014 года работают одни и те же платформы. Лишь платформа PSP прекратила свою работу. Поэтому актуальным периодом буду считать 2014-2016 года.

```python
# Перезапишу датафрейм, оставлю данные только с 2014 года без платформы PSP.
# Дальнейший анализ буду проводить по обновленному датафрейму
df = df.query('year_of_release>=2014 and platform!="PSP"')
```

### Продажи на платформах

### Влияние отзывов на продажи

```python
# Построю графики и посчитаю корелляцию
y = 'total_sales'
for platform, games_on_pl in df.groupby('platform'):
    print('='*60)
    print(platform)
    # Посчитаю сколько в колонке не пустых отзывов пользователей
    not_user = len(games_on_pl[games_on_pl['user_score'].notna() == True])
    # Посчитаю сколько в колонке не пустых отзывов критиков
    not_critic = len(games_on_pl[games_on_pl['critic_score'].notna() == True])
    # Задам количество не пустых значений для вывода диаграммы и расчета корреляции
    sum_not = 10
    if not_user > sum_not:
        games_on_pl.plot(kind='scatter', x='user_score', y=y, color='b')
        display(games_on_pl[['user_score', y]].corr()[y])
        plt.show()
    else:
        print('Для платформы',platform, 'не хватает данных для построения диаграммы и расчета корреляции отзывов пользователей')
    if not_critic > sum_not:
        games_on_pl.plot(kind='scatter', x='critic_score', y=y, color='b')
        display(games_on_pl[['critic_score', y]].corr()[y])
        plt.show()
    else:
        print('Для платформы',platform, 'не хватает данных для построения диаграммы и расчета корреляции отзывов критиков')
```

Судя по коэффициенту корреляции есть связь между оценками критиков и пользователей.   
В целом оценки пользователей оказываю меньшее влияние на продажи, чем оценки критиков. Но оценки критиков все-таки не сильно влияют на продажи. Так же есть сильная связь между оценками критиков и пользователей.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC

```python
# Рассчитаю средние значения
xone_mean = df.query('platform == "XOne"')['user_score'].mean()
pc_mean = df.query('platform == "PC"')['user_score'].mean()

print('Средний пользовательский рейтинг Xbox One:', xone_mean)
print('Средний пользовательский рейтинг PC:', pc_mean)
```

```python
# Сформирую данные для проверки гипотезы
xone = df.query('platform == "XOne"')['user_score'].dropna()
pc = df.query('platform == "PC"')['user_score'].dropna()

#Рассчитаю дисперсии
print('Дисперсия XOne =', np.var(xone))    
print('Дисперсия PC =',  np.var(pc))
```

Сформулирую нулевую и альтернативную гипотезы:
- Н0 = Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
- Н1 = Средние пользовательские рейтинги платформ Xbox One и PC разные

Т.к. я проверяю гипотезу о равенстве среднего двух генеральных выборок, то буду использовать метод scipy.stats.ttest_ind (array1, array2, equal_var). equal_var=False т.к. дисперсии не равны между собой.

Значение критического уровня вероятности определю равным 5%.

```python
alpha = .05

results = st.ttest_ind(
    xone, 
    pc, 
    equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод:**
Нельзя отвергнуть гипотезу о равенстве средних пользовательских рейтингов у платформ Xbox One и PC , а значит средние пользовательские рейтинги с большой вероятность равны.
Ранее выяснила, что средний пользовательский рейтинг для PC 6.3, а для Xbox One 6.6, что подтверждает проверку.

### Средние пользовательские рейтинги жанров Actoin и Sports

```python
#Рассчитаю средние значения
action_mean = df.query('genre == "Action"')['user_score'].mean()
sports_mean = df.query('genre == "Sports"')['user_score'].mean()

print('Средний пользовательский рейтинг Action:', action_mean)
print('Средний пользовательский рейтинг Sports:', sports_mean)
```

```python
# Сформирую данные для проверки гипотезы
action = df.query('genre == "Action"')['user_score'].dropna()
sports = df.query('genre == "Sports"')['user_score'].dropna()

#Рассчитаю дисперсии
print('Дисперсия Action =', np.var(action))    
print('Дисперсия Sports =',  np.var(sports))
```

Сформулирую нулевую и альтернативную гипотезы:
- Н0 = Средние пользовательские рейтинги жанров Action и Sports одинаковые
- Н1 = Средние пользовательские рейтинги платформ Action и Sports разные

Т.к. я проверяю гипотезу о равенстве среднего двух генеральных выборок, то буду использовать метод scipy.stats.ttest_ind (array1, array2, equal_var). equal_var=False т.к. дисперсии не равны между собой.

Значение критического уровня вероятности определю равным 5%.

```python
alpha = .05

results = st.ttest_ind(
    action, 
    sports, 
    equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод:**
Отвергаем нулевую гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports. Значит средние пользовательские рейтинги с большой вероятность отличаются.
Так как гипотеза двухсторонняя, то сама проверка не даёт понять в какую именно сторону оно отличается.
Но как мы уже выяснили ранее: средний пользовательский рейтинг жанра Action 6.76, а для Sports 5.22.
